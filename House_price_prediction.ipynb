```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import sklearn.datasets
from sklearn import metrics
```

```python
house_price_dataset= pd.read_csv('https://lib.stat.cmu.edu/datasets/boston', sep="\s+", skiprows=22, header=None)
```

```python
print(house_price_dataset)
```

```python
data = np.hstack([house_price_dataset.values[::2, :], house_price_dataset.values[1::2, :2]])
```

```python
print(data)
```

```python
house_price_dataframe=pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
```

```python
house_price_dataframe.head()
```

```python
target = house_price_dataset.values[1::2, 2]
```

```python
house_price_dataframe['PRICE']=target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
correlation=house_price_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X=house_price_dataframe.drop(['PRICE'], axis=1)
Y=house_price_dataframe['PRICE']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X.shape)
```

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
score_1=metrics.r2_score(Y_train,training_data_prediction)
```

```python
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print(score_1)
print(score_2)
```

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
test_data_prediction=model.predict(X_test)
```

```python
score_1=metrics.r2_score(Y_test,test_data_prediction)
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)
print(score_1)
print(score_2)
```
